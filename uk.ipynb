```python
import requests
from bs4 import BeautifulSoup
import urllib.request
from urllib.request import urlopen
url="https://www.collegedekho.com/btech-colleges-in-india/"
page=urllib.request.urlopen(url)
page_soup=BeautifulSoup(page,'html.parser')
img_items=page_soup.find('div',{'class':'collegeinfo'})
img_div=img_items.find_all(class_="collegeLogo")

for img in img_div:
    image=img.find("img")
    print(image)
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.collegedekho.com/btech-colleges-in-india/"

response = requests.get(url)
page_soup = BeautifulSoup(response.content, 'html.parser')
img_items = page_soup.find_all('div', {'class': 'collegeLogo'})

for img_item in img_items:
    img_src = img_item.find('img')['src']
    img_content = img_item.find('img')['alt']
    print("Image Source:", img_src)
    print("Image Content:", img_content)
    print("---------------------------------")
```
